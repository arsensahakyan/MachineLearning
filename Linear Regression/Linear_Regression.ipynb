```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"housing_prices.csv") 
```

```python
norm_data = data.apply(lambda x: x / x.max(), axis=0)
norm_data
```

```python
training_set = norm_data.iloc[:41]
```

```python
testing_set = norm_data.iloc[41:]
```

```python
training_set.insert(0, 'X0', [1] * 41)
training_set
```

```python
testing_set.insert(0, "X0", [1] * 6)
```

```python
thetas = [0, 0, 0]

def hypothesis_function(thetas, i):
    res = 0
    for j in range(3):
        res += thetas[j] * training_set.iloc[i, j]
    return res
```

```python
def derivative_calculation(j, thetas):
    res = 0
    for i in range(41):
        res += (hypothesis_function(thetas, i) - training_set.iloc[i, 3]) * training_set.iloc[i, j]
    return res
```

```python
alpha = 0.001
```

```python
thetas = [0, 0, 0]
def linreg():
    global thetas
    for _ in range(1000):
        temp_thetas = thetas[:]
        for j in range(3):
            temp_thetas[j] = temp_thetas[j] - alpha * derivative_calculation(j, thetas)
        thetas = temp_thetas[:]
    return thetas
print(linreg())
```
