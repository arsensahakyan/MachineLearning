```python
import numpy as np
dx = 0.1; dy = 0.1
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)
def f(x,y):  
        return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
C = plt.contour(X,Y,f(X,Y),8,colors='black') #contour
plt.contourf(X,Y,f(X,Y),8,cmap=plt.cm.hot) #cmap - color
plt.clabel(C, fontsize=10) #values of f(x,y) on chart
plt.colorbar()
```

## 3D Surfaces

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
ax.plot_surface(X,Y,f(X,Y), rstride=10, cstride=10)
```
