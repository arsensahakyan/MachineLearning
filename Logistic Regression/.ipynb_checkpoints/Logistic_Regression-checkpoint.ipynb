```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"ClassificationDataset.csv") 
```

```python
norm_data = data.apply(lambda x: x / x.max(), axis=0)
norm_data
```

```python
training_set = norm_data.iloc[:1300]
```

```python
testing_set = norm_data.iloc[1301:]
```

```python
training_set.insert(0, 'X0', [1] * 1300)
training_set
```

```python
testing_set.insert(0, "X0", [1] * 71)
testing_set
```

```python
thetas = [0, 0, 0, 0, 0]

def hypothesis_function(thetas, i):
    res = 0
    for j in range(5):
        res += thetas[j] * training_set.iloc[i, j]
    final_res = 1 / (1 + math.exp(-res))
    return final_res
```

```python
def derivative_calculation(j, thetas):
    res = 0
    for i in range(1300):
        res += (hypothesis_function(thetas, i) - training_set.iloc[i, 3]) * training_set.iloc[i, j]
    return res
```

```python
alpha = 0.01
```

```python
thetas = [0, 0, 0, 0, 0]
def logreg():
    global thetas
    for _ in range(50):
        temp_thetas = thetas[:]
        for j in range(5):
            temp_thetas[j] = temp_thetas[j] - alpha * derivative_calculation(j, thetas)
        thetas = temp_thetas[:]
    return thetas
print(logreg())
```

```python
def hypothesis_function_p(thetas, i):
    res = 0
    for j in range(5):
        res += thetas[j] * norm_data.iloc[i, j]
    final_res = 1 / (1 + math.exp(-res))
    return final_res
print(hypothesis_function_p(thetas, 1369))
```
